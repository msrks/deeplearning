```python
import numpy as np
import  matplotlib.pyplot as plt
%matplotlib inline
from  matplotlib.cm import Vega10 # http://matplotlib.org/examples/color/colormaps_reference.html
import seaborn as sns
sns.set(style='white')

from sklearn.datasets import load_digits
from sklearn.manifold import TSNE
```

```python
X, y = load_digits(return_X_y=True)
model = TSNE(n_components=2)
tsne_result = model.fit_transform(X) 
```

```python
fig = plt.figure()
im = plt.scatter(tsne_result[:,0], tsne_result[:,1], c=y, cmap=Vega10, s=1.5)
fig.colorbar(im)
```
